```python
import json
import numpy as np

file_path = "../../experiments/gpt-oss-120b_03_fixed_date/population_5_5.json"

with open(file_path, 'r') as f:
    data = json.load(f)

activities_by_person = [p['identity'].get("activities", []) for p in data]
activities_by_person

for acts in activities_by_person:
    assert acts[0]['start_time'] == -1.0, f"Person {activities_by_person.index(acts)} has invalid start time: {acts[0]['start_time']}"
    start_time_arr = np.array([act['start_time'] for act in acts[1:]])
    assert np.all(np.diff(start_time_arr) >= 0), f"Person {activities_by_person.index(acts)} has non-monotonic start times"
    assert np.all(start_time_arr >= 2*3600), f"Person {activities_by_person.index(acts)} has start times less than 2 hours"
    assert np.all(start_time_arr <= 22*3600), f"Person {activities_by_person.index(acts)} has start times greater than 22 hours"
```
